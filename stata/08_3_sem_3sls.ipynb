# Estimation of 3SLS

```python
%%capture
import stata_setup
stata_setup.config("C:\Program Files\StataNow19", "mp")
```

```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/8-sem.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # ラグ変数の作成
tsset year
gen l_gdp  = L.gdp
gen l_m    = L.m
gen l_i    = L.i
gen l_gov  = L.gov
gen l_tb   = L.tb
```

```python
%%stata
* # 欠損行を削除
drop if missing(l_gdp, l_m, l_i, l_gov, l_tb, gdp, m, i, gov, tb)
```

```python
%%stata
* # 3SLSの推定
reg3 (gdp m i gov) (m gdp tb), endog(m gdp) exog(i gov tb l_gdp l_m l_i l_gov l_tb) 3sls
```
