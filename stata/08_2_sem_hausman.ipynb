```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/8-sem.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # ラグ変数の作成
tsset year
gen l_gdp  = L.gdp
gen l_m    = L.m
gen l_i    = L.i
gen l_gov  = L.gov
gen l_tb   = L.tb
```

```python
%%stata
* # 欠損行を削除
drop if missing(l_gdp, l_m, l_i, l_gov, l_tb, gdp, m, i, gov, tb)
```

```python
%%stata
* # OLS
regress m i gov tb l_gdp l_m l_i l_gov l_tb
```

```python
%%stata
* # 残差を保存
predict residm, resid
```

```python
%%stata
* # GDPをMとresidMで説明
regress gdp m i gov residm
```
