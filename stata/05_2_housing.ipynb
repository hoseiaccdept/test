# Testing and correcting for Heteroskedasticity
住宅支出（HEXP）を所得（INC）で説明する単回帰モデルを例に、以下の3ステップを行う  
1. 通常のOLS推定  
2. White検定による不均一分散の検定  
3. Whiteロバスト標準誤差による修正推定  

```python
%%capture
import stata_setup
stata_setup.config("C:\Program Files\StataNow19", "mp")
```

```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/5-housing.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # 通常のOLS
regress hexp inc
```

```python
%%stata
* # White検定による不均一分散の検定
estat imtest, white
```

```python
%%stata
* # Whiteの頑健標準誤差を用いた推定量
regress hexp inc, vce(robust)
```
