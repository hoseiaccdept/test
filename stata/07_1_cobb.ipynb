# Controlling for fixed effects
固定効果（年・業種）をダミーで入れた回帰分析を行います。

```python
%%capture
import stata_setup
stata_setup.config("C:\Program Files\StataNow19", "mp")
```

```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/7-cobb.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # 対数変換
gen ln_q  = log(q)
gen ln_k  = log(k)
gen ln_l  = log(l)
gen ln_it = log(it)
```

```python
%%stata
* # 年のダミー（1990を基準：D91, D92）
gen d91 = year == 1991
gen d92 = year == 1992
```

```python
%%stata
* # 業種ダミー（10を基準：IND2 for 20, IND3 for 30）
gen ind2 = industry == 20
gen ind3 = industry == 30
```

```python
%%stata
* # 回帰分析
regress ln_q ln_k ln_l ln_it d91 d92 ind2 ind3
```
