# Ordinary least squares estimation
Income(X)とConsumption(Y)の関係を単回帰モデルで分析

```python
%%capture
import stata_setup
stata_setup.config("C:\Program Files\StataNow19", "mp")
```

```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/2-cons-sample.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # 回帰モデル
regress consumption income
```

```python
%%stata
# 変数作成
gen X_STAR = income * 1000
gen Y_STAR = consumption * 1000
```

```python
%%stata
* # 回帰モデル
regress Y_STAR X_STAR
```
