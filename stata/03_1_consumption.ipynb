# Multiple linear regression models
消費を目的変数として、所得、家族人数、年齢を説明変数とする重回帰分析を行う

```python
%%capture
import stata_setup
stata_setup.config("C:\Program Files\StataNow19", "mp")
```

```python
%%stata
* # CSVファイルの読み込み
import delimited "../data/3-consumption.csv", clear
```

```python
%%stata
* # データを確認（先頭5件）
list in 1/5
```

```python
%%stata
* # 重回帰分析
regress y x1 x2 x3
```


$$
H_0:\ \beta_2 = \beta_3 = 0
$$

```python
%%stata
* # Wald検定(F-statistic)
quietly regress y x1 x2 x3
test (x2 = 0) (x3 = 0)
```

```python
%%stata
* # Wald検定(Chi-statistic)
quietly glm y x1 x2 x3, family(gaussian) link(identity)
test (x2 = 0) (x3 = 0)
```

$$
H_0:\ \beta_1 = 3\beta_2 + \beta_3
$$

```python
%%stata
* # Wald検定(F-statistic)
quietly regress y x1 x2 x3
test x1 = 3*x2 + x3
```

```python
%%stata
* # Wald検定(F-statistic)
quietly glm y x1 x2 x3, family(gaussian) link(identity)
test x1 = 3*x2 + x3
```
