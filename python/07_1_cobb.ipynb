# Controlling for fixed effects
固定効果（年・業種）をダミーで入れた回帰分析を行います。

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/7-cobb.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 対数変換
data["LN_Q"] = np.log(data["Q"])
data["LN_K"] = np.log(data["K"])
data["LN_L"] = np.log(data["L"])
data["LN_IT"] = np.log(data["IT"])
```

```python
# 年のダミー（1990を基準：D91, D92）
data["D91"] = (data["Year"] == 1991).astype(int)
data["D92"] = (data["Year"] == 1992).astype(int)
```

```python
# 業種ダミー（10を基準：IND2 for 20, IND3 for 30）
data["IND2"] = (data["Industry"] == 20).astype(int)
data["IND3"] = (data["Industry"] == 30).astype(int)
```

```python
# 回帰分析
X = data[['LN_K', 'LN_L', 'LN_IT', 'D91', 'D92', 'IND2', 'IND3']]
X = sm.add_constant(X)  # 定数項を追加
y = data['LN_Q']

model = sm.OLS(y, X).fit()
```

```python
# 結果の出力
print(model.summary())
```
