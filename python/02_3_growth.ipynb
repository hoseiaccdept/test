# Log-Linear models
年次データYの自然対数を取って、時間tに対する成長率を推定

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import warnings
# n<20 の場合の kurtosistest 警告を無視
warnings.filterwarnings("ignore", message="kurtosistest only valid for n>=20")
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/2-growth.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 対数変換
data["lnY"] = np.log(data["Y"])
```

```python
# 回帰分析
X = sm.add_constant(data["t"])
y = data["lnY"]

model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```
