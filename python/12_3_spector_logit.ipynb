# Logit model

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/12-Spector.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# Logit回帰モデル
X = sm.add_constant(data[['GPA','TUCE','PSI']])
y = data['GRADE']

model_logit = sm.Logit(y, X).fit()
```

```python
# 結果の出力
print(model_logit.summary())
```
