# Testing and correcting for Heteroskedasticity
住宅支出（HEXP）を所得（INC）で説明する単回帰モデルを例に、以下の3ステップを行う  
1. 通常のOLS推定  
2. White検定による不均一分散の検定  
3. Whiteロバスト標準誤差による修正推定  

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white
from statsmodels.stats.api import het_breuschpagan
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/5-housing.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 通常のOLS
X = sm.add_constant(data["INC"])
y = data["HEXP"]
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```

```python
# White検定による不均一分散の検定
white_test = het_white(model.resid, model.model.exog)
```

```python
labels = ["LM Statistic", "LM-Test p-value", "F-Statistic", "F-Test p-value"]
print(dict(zip(labels, white_test)))
```

```python
# Whiteの頑健標準誤差を用いた推定量
robust_result = model.get_robustcov_results(cov_type='HC1')
```

```python
print(robust_result.summary())
```
