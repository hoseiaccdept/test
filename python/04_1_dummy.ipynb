# Dummy variable Regression
消費を目的変数として、世帯主が男性か女性か（ダミー変数）と、所得（数量変数）の両方を説明変数とする回帰分析

```python
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/4-dummy.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# モデル1: ダミー変数のみ
X = sm.add_constant(data["MALE"])
y = data["Y"]
model1 = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model1.summary())
```

```python
# モデル2
X = data[["MALE", "X"]]
y = data["Y"]
X = sm.add_constant(X)
model2 = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model2.summary())
```
