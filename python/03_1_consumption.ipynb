# Multiple linear regression models
消費を目的変数として、所得、家族人数、年齢を説明変数とする重回帰分析を行う

```python
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/3-consumption.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 重回帰分析
X = data[["X1", "X2", "X3"]]
y = data["Y"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```

```python
# Chi-statistic算出のため
glm_model = sm.GLM(y, X, family=sm.families.Gaussian()).fit()
```


$$
H_0:\ \beta_2 = \beta_3 = 0
$$

```python
# Wald検定(F-statistic)
print(model.f_test("X2 = 0, X3 = 0"))
```

```python
# Wald検定(Chi-statistic)
print(glm_model.wald_test("X2 = 0, X3 = 0", scalar=True))
```

$$
H_0:\ \beta_1 = 3\beta_2 + \beta_3
$$

```python
# Wald検定(F-statistic)
print(model.f_test("X1 = 3*X2 + X3"))
```

```python
# Wald検定(Chi-statistic)
print(glm_model.wald_test("X1 = 3*X2 + X3"))
```
