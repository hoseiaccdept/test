# Estimation of 2SLS
目的変数： GDP  
内生変数： M  
外生変数： I, GOV  
操作変数（IVs）： I, GOV, TB, L_GDP, L_M, L_I, L_GOV, L_TB  

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from linearmodels.iv import IV2SLS
# インストールコマンド例: pip install linearmodels
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/8-sem.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# ラグ変数の作成
data = data.sort_values('Year')
data['L_GDP'] = data['GDP'].shift(1)
data['L_M']   = data['M'].shift(1)
data['L_I']   = data['I'].shift(1)
data['L_GOV'] = data['GOV'].shift(1)
data['L_TB']  = data['TB'].shift(1)
```

```python
# 欠損行を削除
data = data.dropna(subset=['L_GDP', 'L_M', 'L_I', 'L_GOV', 'L_TB', 'GDP', 'M', 'I', 'GOV', 'TB'])
```

```python
# 2SLS推定
formula = "GDP ~ 1 + I + GOV + [M ~ TB + L_GDP + L_M + L_I + L_GOV + L_TB]"
model = IV2SLS.from_formula(formula, data).fit(cov_type='unadjusted', debiased=True)
```

```python
# 結果表示
print(model)
```
