# t-test
Processの平均が224と異なるかをt検定する

```python
import pandas as pd
from scipy import stats
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/1-Ex 1-4.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# t検定
t_result = stats.ttest_1samp(data["Process"], popmean=224)
```

```python
# 結果の表示
print(t_result)
```
