# Estimation of 3SLS

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from linearmodels.system import IV3SLS
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/8-sem.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# ラグ変数の作成
data = data.sort_values('Year')
data['L_GDP'] = data['GDP'].shift(1)
data['L_M']   = data['M'].shift(1)
data['L_I']   = data['I'].shift(1)
data['L_GOV'] = data['GOV'].shift(1)
data['L_TB']  = data['TB'].shift(1)
```

```python
# 欠損行を削除
data = data.dropna(subset=['L_GDP', 'L_M', 'L_I', 'L_GOV', 'L_TB', 'GDP', 'M', 'I', 'GOV', 'TB'])
```

```python
# 各方程式の定義
formulas = {
    'GDPeq': 'GDP ~ 1 + I + GOV + [M ~ TB + L_GDP + L_M + L_I + L_GOV + L_TB]',
    'Meq':   'M ~ 1 + TB + [GDP ~ I + GOV + L_GDP + L_M + L_I + L_GOV + L_TB]'
}
```

```python
# 3SLSの推定
model_3sls = IV3SLS.from_formula(formulas, data=data).fit(cov_type='unadjusted', debiased=True)
```

```python
# 結果表示
print(model_3sls.summary)
```
