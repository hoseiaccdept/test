# Multicollinearity

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_white
from statsmodels.stats.api import het_breuschpagan
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/5-Multicollinearity.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 単回帰分析
X = sm.add_constant(data["X1"])
y = data["Y"]
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```

```python
# 重回帰分析
X = data[["X1", "X2"]]
y = data["Y"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```
