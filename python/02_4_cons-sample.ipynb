# Ordinary least squares estimation
Income(X)とConsumption(Y)の関係を単回帰モデルで分析

```python
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/2-cons-sample.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 回帰モデル
X = data["Income"]
y = data["Consumption"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```

```python
# 回帰モデル
X_star = data["Income"] * 1000
y_star = data["Consumption"] * 1000
X_star = sm.add_constant(X_star)
model_star = sm.OLS(y_star, X_star).fit()
```

```python
# 結果の表示
print(model_star.summary())
```
