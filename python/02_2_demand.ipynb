# Log-log linear models
価格と需要の関係を、両方の変数を対数変換して回帰分析

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import warnings
# n<20 の場合の kurtosistest 警告を無視
warnings.filterwarnings("ignore", message="kurtosistest only valid for n>=20")
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/2-demand.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 対数変換
data["lnP"] = np.log(data["P"])
data["lnQ"] = np.log(data["Q"])
```

```python
# 回帰分析
X = sm.add_constant(data["lnP"])  # 定数項を追加
y = data["lnQ"]
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```
