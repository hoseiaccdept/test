# Probit model


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/12-Spector.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# Probit回帰モデル
X = data[['GPA', 'TUCE', 'PSI']]
X = sm.add_constant(X)
y = data['GRADE']

model_probit = sm.GLM(y, X, family=sm.families.Binomial(link=sm.families.links.probit())).fit()
```

```python
# 結果の出力
print(model_probit.summary())
```

```python
# ↓だと数値がことなる
model_probit = sm.Probit(y, X).fit()
print(model_probit.summary())
```

```python
# Probit モデル推定
model = sm.Probit(y, X).fit(cov_type='HC3')

print(model.summary())
```
