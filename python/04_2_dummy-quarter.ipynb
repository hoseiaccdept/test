# Dummy variable Regression
目的変数：四半期RGDP成長率  
説明変数：金利、政府支出成長、四半期ダミー

```python
import pandas as pd
import statsmodels.api as sm
```

```python
# CSVファイルの読み込み
data = pd.read_csv("../data/4-dummy-quarter.csv")
```

```python
# データを確認（先頭5件）
data.head()
```

```python
# 四半期ダミーを作成（Q4を基準）
data["D1"] = (data["QT"] == "Q1").astype(int)
data["D2"] = (data["QT"] == "Q2").astype(int)
data["D3"] = (data["QT"] == "Q3").astype(int)
```

```python
# 重回帰分析
X = data[["INTEREST", "RGEXP", "D1", "D2", "D3"]]
y = data["RGDP"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
# 結果の表示
print(model.summary())
```
