# Hausman test

```r
# CSVファイルの読み込み
data <- read.csv("../data/8-sem.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# ラグ変数の作成
data <- data[order(data$Year), ]
data$L_GDP <- c(NA, head(data$GDP, -1))
data$L_M   <- c(NA, head(data$M, -1))
data$L_I  <- c(NA, head(data$I, -1))
data$L_GOV <- c(NA, head(data$GOV, -1))
data$L_TB  <- c(NA, head(data$TB, -1))
```

```r
# OLS
model_stage1 <- lm(M ~ I + GOV + TB + L_GDP + L_M + L_I + L_GOV + L_TB, data = data, na.action = na.exclude)
```

```r
# 結果の出力
summary(model_stage1)
```

```r
# 残差を保存
data$residM <- residuals(model_stage1)
```

```r
# GDPをMとresidMで説明
model_stage2 <- lm(GDP ~ M + I + GOV + residM, data = data)
```

```r
# 結果の出力
summary(model_stage2)
```
