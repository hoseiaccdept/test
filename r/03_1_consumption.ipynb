# Multiple linear regression models
消費を目的変数として、所得、家族人数、年齢を説明変数とする重回帰分析を行う

Rの標準関数に加えて、`car` パッケージを使用します。

```r
# 必要パッケージ
if (!require(car)) install.packages("car")
library(car)
```

```r
# CSVファイルの読み込み
data <- read.csv("../data/3-consumption.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# 重回帰分析
model <- lm(Y ~ X1 + X2 + X3 , data = data)
```

```r
# 結果の表示
summary(model)
```


$$
H_0:\ \beta_2 = \beta_3 = 0
$$

```r
# Wald検定(F-statistic)
linearHypothesis(model, c("X2 = 0", "X3 = 0"), test = "F")
```

```r
# Wald検定(Chi-statistic)
linearHypothesis(model, c("X2 = 0", "X3 = 0"), test = "Chisq")
```

$$
H_0:\ \beta_1 = 3\beta_2 + \beta_3
$$

```r
# Wald検定(F-statistic)
linearHypothesis(model, "X1 = 3*X2 + X3", test = "F")
```

```r
# Wald検定(F-statistic)
linearHypothesis(model, "X1 = 3*X2 + X3", test = "Chisq")
```
