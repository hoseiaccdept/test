# Tobit model


Rの標準関数に加えて、`AER` パッケージを使用します。

```r
if (!require(AER)) install.packages("AER")
library(AER)
```

```r
# CSVファイルの読み込み
data <- read.csv("../data/12-censored.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# トービットモデル
fit <- tobit(Y ~ Z1 + Z2, left = 0, data = data)
```

```r
# 結果の出力
summary(fit)
```
