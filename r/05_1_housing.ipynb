# Testing and correcting for Heteroskedasticity
住宅支出（HEXP）を所得（INC）で説明する単回帰モデルを例に、以下の3ステップを行います。  
1. 通常のOLS推定  
2. White検定による不均一分散の検定  
3. Whiteロバスト標準誤差による修正推定  

Rの標準関数に加えて、検定に `lmtest`、修正推定に `sandwich` パッケージを使用します。

```r
if (!require(lmtest)) install.packages("lmtest")
library(lmtest)
```

```r
if (!require(sandwich)) install.packages("sandwich")
library(sandwich)
```

```r
# CSVファイルの読み込み
data <- read.csv("../data/5-housing.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# 通常のOLS
model <- lm(HEXP ~ INC, data = data)
```

```r
# 結果の表示
summary(model)
```

```r
# White検定による不均一分散の検定
bptest(model, ~ fitted(model) + I(fitted(model)^2))
```

```r
coeftest(model, vcov = vcovHC(model, type = "HC1"))
```
