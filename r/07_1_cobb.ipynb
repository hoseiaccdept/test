# Controlling for fixed effects
固定効果（年・業種）をダミーで入れた回帰分析を行います。

```r
# CSVファイルの読み込み
data <- read.csv("../data/7-cobb.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# 対数変換
data$LN_Q  <- log(data$Q)
data$LN_K  <- log(data$K)
data$LN_L  <- log(data$L)
data$LN_IT <- log(data$IT)
```

```r
# 年のダミー（1990を基準：D91, D92）
data$D91 <- as.integer(data$Year == 1991)
data$D92 <- as.integer(data$Year == 1992)
```

```r
# 業種ダミー（10を基準：IND2 for 20, IND3 for 30）
data$IND2 <- as.integer(data$Industry == 20)
data$IND3 <- as.integer(data$Industry == 30)
```

```r
# 回帰分析
model <- lm(LN_Q ~ LN_K + LN_L + LN_IT + D91 + D92 + IND2 + IND3, data = data)
```

```r
# 結果の出力
summary(model)
```
