# Log-log linear models
価格と需要の関係を、両方の変数を対数変換して回帰分析

```r
# CSVファイルの読み込み
data <- read.csv("../data/2-demand.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# 対数変換
data$lnP <- log(data$P)
data$lnQ <- log(data$Q)
```

```r
# 回帰分析
model <- lm(lnQ ~ lnP, data = data)
```

```r
# 結果の表示
summary(model)
```
