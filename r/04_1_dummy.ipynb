# Dummy variable Regression
消費を目的変数として、世帯主が男性か女性か（ダミー変数）と、所得（数量変数）の両方を説明変数とする回帰分析を行います。

```r
# CSVファイルの読み込み
data <- read.csv("../data/4-dummy.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# モデル1: ダミー変数のみ
model1 <- lm(Y ~ MALE , data = data)
```

```r
# 結果の表示
summary(model1)
```

```r
# モデル2
model2 <- lm(Y ~ MALE + X , data = data)
```

```r
# 結果の表示
summary(model2)
```
