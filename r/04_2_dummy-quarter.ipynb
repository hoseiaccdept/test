# Dummy variable Regression
目的変数：四半期RGDP成長率  
説明変数：金利、政府支出成長、四半期ダミー

```r
# CSVファイルの読み込み
data <- read.csv("../data/4-dummy-quarter.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# 四半期ダミーを作成（Q4を基準）
data$D1 <- as.integer(data$QT == 'Q1')
data$D2 <- as.integer(data$QT == 'Q2')
data$D3 <- as.integer(data$QT == 'Q3')
```

```r
# 重回帰分析
model <- lm(RGDP ~ INTEREST + RGEXP + D1 + D2 + D3, data = data)
```

```r
# 結果の表示
summary(model)
```
