# Estimation of 2SLS
目的変数： GDP  
内生変数： M  
外生変数： I, GOV  
操作変数（IVs）： I, GOV, TB, L_GDP, L_M, L_I, L_GOV, L_TB  

Rの標準関数に加えて、`AER` パッケージを使用します。

```r
if (!require(AER)) install.packages("AER")
library(AER)
```

```r
# CSVファイルの読み込み
data <- read.csv("../data/8-sem.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# ラグ変数の作成
data <- data[order(data$Year), ]
data$L_GDP <- c(NA, head(data$GDP, -1))
data$L_M   <- c(NA, head(data$M, -1))
data$L_I  <- c(NA, head(data$I, -1))
data$L_GOV <- c(NA, head(data$GOV, -1))
data$L_TB  <- c(NA, head(data$TB, -1))
```

```r
# 2SLS 推定
iv2sls <- ivreg(GDP ~ M + I + GOV | I + GOV + TB + L_GDP + L_M + L_I + L_GOV + L_TB, data = data)
```

```r
# 結果の出力
summary(iv2sls)
```
