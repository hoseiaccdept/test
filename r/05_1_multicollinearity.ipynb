# Multicollinearity

```r
# CSVファイルの読み込み
data <- read.csv("../data/5-Multicollinearity.csv")
```

```r
# データを確認（先頭5件）
head(data, 5)
```

```r
# OLS
model <- lm(Y ~ X1, data = data)
```

```r
# 結果の表示
summary(model)
```

```r
# 重回帰分析
model <- lm(Y ~ X1 + X2, data = data)
```

```r
# 結果の表示
summary(model)
```
